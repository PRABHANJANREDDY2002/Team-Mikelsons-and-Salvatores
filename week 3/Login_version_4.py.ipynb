```python
from tkinter import *
from functools import partial
from PIL import ImageTk, Image


def log_in(master):
    if len(master)==0:
        loginwn=Tk()
        loginwn.geometry("100x100")
        loginwn.configure(bg="red")
        fr1=Frame(loginwn,bg="blue")
        l_title=Message(loginwn,text="Invalid",relief="raised",width=200,padx=60,pady=0,fg="white",bg="black",justify="center",anchor="center")
        l_title.config(font=("Courier","10","bold"))
        l_title.pack(side="top")
    else:
        loginwn=Tk()
        loginwn.geometry("300x300")
        loginwn.title("Log in")
        loginwn.configure(bg="green")
        fr1=Frame(loginwn,bg="blue")
        l_title=Message(loginwn,text="welcome",relief="raised",width=2000,padx=600,pady=0,fg="white",bg="black",justify="center",anchor="center")
        l_title.config(font=("Courier","50","bold"))
        l_title.pack(side="top")

    
def Create():
    crwn=Tk()  
    crwn.geometry("800x800")
    crwn.title("Create Account")
    crwn.configure(bg="orange")
    fr1=Frame(crwn,bg="blue")
    l_title=Message(crwn,text="OUR BANK",relief="raised",width=2000,padx=600,pady=0,fg="white",bg="black",justify="center",anchor="center")
    l_title.config(font=("Courier","50","bold"))
    l_title.pack(side="top")
    l1=Label(crwn,text="Enter Name:",relief="raised")
    l1.pack(side="top")
    e1=Entry(crwn)
    e1.pack(side="top")
    l2=Label(crwn,text="Enter opening credit:",relief="raised")
    l2.pack(side="top")
    e2=Entry(crwn)
    e2.pack(side="top")
    l3=Label(crwn,text="Enter desired PIN:",relief="raised")
    l3.pack(side="top")
    e3=Entry(crwn,show="*")
    e3.pack(side="top")
    b=Button(crwn,text="Submit",command=lambda: write(crwn,e1.get().strip(),e2.get().strip(),e3.get().strip()))
    b.pack(side="top")
    crwn.bind("<Return>",lambda x:write(crwn,e1.get().strip(),e2.get().strip(),e3.get().strip()))
    return

def main_menu():
    tkWindow = Tk()  
    tkWindow.geometry("800x800")  
    tkWindow.title('Banking System')
    
    frame1 = Frame(tkWindow , width=600, height=600)
    frame1.pack()
    frame1.place(anchor='center', relx=0.5, rely=0.5)
    """canvas = Canvas(tkWindow,width=600, height=800, bg='blue')
    canvas.pack(expand=YES, fill=BOTH)
    image = ImageTk.PhotoImage(file="bankii.png")

    canvas.create_image(0, 0, image=image, anchor=NW)
    """



    frame = Frame(tkWindow,bd=5,highlightthickness=2,highlightbackground="black",padx=50,pady=50)

    frame.place(relx=.5, rely=.5,anchor= CENTER)
    label2 = Label( frame1, text = "Online Banking System",bg = "#88cffa").place(height=50,width=300, x=190, y=110)

    usernameLabel = Label(frame1, text="Account Number").place(height=50, x=190, y=200)
    username = StringVar()
    usernameEntry = Entry(frame1, textvariable=username).place(height=25,width=190, x=290, y=210)

    passwordLabel = Label(frame1,text="Password").place(height=50, x=230, y=250)
    password = StringVar()
    passwordEntry = Entry(frame1, textvariable=password, show='*').place(height=25,width=190, x=290, y=260)  

    loginButton = Button(frame1, text="Login", command=lambda:log_in(username.get()),bg = "#88cffa").place(height=30,width=150, x=190, y=320)
    creatButton = Button(frame1, text="Create",command=Create,bg = "#88cffa").place(height=30,width=150, x=340, y=320)

    tkWindow.mainloop()
main_menu()
```
