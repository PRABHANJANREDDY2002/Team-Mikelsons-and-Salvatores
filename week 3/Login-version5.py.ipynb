```python
from tkinter import *
from functools import partial
from PIL import ImageTk, Image
```

```python
def log_in(master):
    if len(master)==0:
        loginwn=Tk()
        loginwn.geometry("100x100")
        loginwn.configure(bg="red")
        fr1=Frame(loginwn,bg="blue")
        l_title=Message(loginwn,text="Invalid",relief="raised",width=200,padx=60,pady=0,fg="white",bg="black",justify="center",anchor="center")
        l_title.config(font=("Courier","10","bold"))
        l_title.pack(side="top")
    else:
        loginwn=Tk()
        loginwn.geometry("300x300")
        loginwn.title("Log in")
        loginwn.configure(bg="green")
        fr1=Frame(loginwn,bg="blue")
        l_title=Message(loginwn,text="welcome",relief="raised",width=2000,padx=600,pady=0,fg="white",bg="black",justify="center",anchor="center")
        l_title.config(font=("Courier","50","bold"))
        l_title.pack(side="top")
```

```python
def Create():
    
    crwn=Tk()  
    crwn.geometry("800x800")
    crwn.title("Create Account")
    crwn.configure(bg="orange")
    
    fr1=Frame(crwn,bg="blue")
    
    l_title=Message(crwn,text="OUR BANK",relief="raised",width=2000,padx=600,pady=0,fg="white",bg="black",justify="center",anchor="center")
    l_title.config(font=("Courier","50","bold"))
    l_title.pack(side="top")
    
    l1=Label(crwn,text="Enter Name:",relief="raised",width=45,pady=7,bg = "#88cffa")
    l1.pack(side="top")
    
    e1=Entry(crwn,width=50,borderwidth=5)
    e1.pack(side="top")
    l2=Label(crwn,text="opening credit:",relief="raised",width=45,bg = "#88cffa",pady=7)
    l2.pack(side="top")
    e2=Entry(crwn,width=50,borderwidth=5)

    e2.pack(side="top")
    l3=Label(crwn,text="desired PIN:",relief="raised",width=45,bg = "#88cffa",pady=7)
    l3.pack(side="top")
    e3=Entry(crwn,width=50,borderwidth=5)

    e3.pack(side="top")
    b=Button(crwn,text="Submit",width=45,bg = "green",command=lambda: write(crwn,e1.get().strip(),e2.get().strip(),e3.get().strip()))
    b.pack(side="top")
    crwn.bind("<Return>",lambda x:write(crwn,e1.get().strip(),e2.get().strip(),e3.get().strip()))
    return
```

```python
def main_menu():
    def resize_bg(event):
        global bgg, resized, bg2
        # open image to resize it
        bgg = Image.open("bankimage1.png")
        # resize the image with width and height of root
        resized = bgg.resize((event.width, event.height),Image.ANTIALIAS)
    
        bg2 = ImageTk.PhotoImage(resized)
        canvas.create_image(0, 0, image=bg2, anchor='nw')
        
    tkWindow = Tk()
    tkWindow.geometry("800x800")  
    tkWindow.title('Banking System')
    
    canvas = Canvas(tkWindow,width=600, height=800)
    canvas.pack(expand=YES, fill=BOTH)
    image = ImageTk.PhotoImage(file="bankimage1.png")

    canvas.create_image(0, 0, image=image, anchor=NW)
    
    frame1 = Frame(tkWindow , width=600, height=600,bd=5,highlightthickness=2,highlightbackground="black",padx=50,pady=50)
    frame1.pack()
    frame1.place(anchor='center', relx=0.5, rely=0.6)
    
    label2 = Label( frame1, text = "Online Banking System",bg = "#88cffa",width=55,height=2).grid(row=0, column=0,columnspan=2)
    
    usernameLabel = Label(frame1, text="Account Number",pady=15).grid(row=1, column=0)
    username = StringVar()
    usernameEntry = Entry(frame1, textvariable=username).grid(row=1, column=1)

    passwordLabel = Label(frame1,text="Password",pady=15).grid(row=2, column=0)
    password = StringVar()
    passwordEntry = Entry(frame1, textvariable=password, show='*').grid(row=2, column=1)  

    loginButton = Button(frame1, text="Login", command=lambda:log_in(username.get()),bg = "#88cffa",width=25,height=1).grid(row=3, column=0)
    creatButton = Button(frame1, text="Create",command=Create,bg = "#88cffa",width=25,height=1).grid(row=3,column=1)
    
    tkWindow.bind("<Configure>",resize_bg)
    tkWindow.mainloop()
    
main_menu()
```
